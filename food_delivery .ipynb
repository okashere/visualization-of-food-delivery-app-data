```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\Lenovo\Desktop\jupyter\zomato\food_delivery.csv', encoding='unicode_escape')
```

```python
#if it would have any null values present then it can be shown by using this, but there is no null value present and this why its all 0
pd.isnull(df).sum()
```

```python
#if there would be any null values present this would be used to drop them
df.dropna(inplace=True)
```

```python
#basic description of the data
df.describe()
```

```python
#coverting the data type of the "rate" column to float and remove the denominator.
def handleRate(value):
    value=str(value).split('/')
    value = value[0];
    return float(value)
df['rate']=df['rate'].apply(handleRate)
print(df.head())
```

```python
#finding types of restaurants
sns.countplot(x=df['listed_in(type)'])
plt.xlabel("Type of restaurant")
```

```python
#the majority of the restaurants falls in the category of dining.
```

```python
# Group by 'listed_in(type)' and sum 'votes'
grouped_data = df.groupby('listed_in(type)')['votes'].sum()

# Convert to DataFrame for plotting
result = pd.DataFrame({'votes': grouped_data})

# Plot the data
plt.plot(result.index, result['votes'], c="green", marker="o")
plt.xlabel("Type of restaurant", c="red", size=20)
plt.ylabel("Votes", c="red", size=20)
plt.xticks(rotation=45)  # Optional: Rotate x-axis labels if they are long
plt.show()
```

```python
# Plot histogram of 'rate' column with 5 bins
plt.hist(df['rate'], bins=5, color='blue', edgecolor='black')
plt.title("Ratings Distribution")
plt.xlabel("Rate")
plt.ylabel("Frequency")
plt.show()
```

```python
# Plot countplot for 'approx_cost(for two people)'
plt.figure(figsize=(8, 6))  # Adjust the size of the figure if needed
sns.countplot(x=df['approx_cost(for two people)'])
plt.title("Count of Approx Cost for Two People")
plt.xlabel("Approx Cost (for two people)")
plt.ylabel("Count")
plt.xticks(rotation=45)  # Rotate x-axis labels if they are long
plt.show()
```

```python
# Plot boxplot of 'rate' grouped by 'online_order'
plt.figure(figsize=(8, 6))  # Adjust the size of the figure if needed
sns.boxplot(x="online_order", y="rate", data=df)
plt.title("Boxplot of Ratings by Online Order")
plt.xlabel("Online Order")
plt.ylabel("Rate")
plt.show()
```

```python
# Create a pivot table for heatmap
pivot_table = df.pivot_table(index="listed_in(type)", columns="online_order", values="rate", aggfunc='mean')
```

```python
# Plot heatmap of the pivot table
plt.figure(figsize=(10, 8))  # Adjust the size of the figure if needed
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt='.1f')
plt.title("Heatmap of Ratings by Listed In (Type) and Online Order")
plt.xlabel("Online Order")
plt.ylabel("Listed In (Type)")
plt.show()
```
